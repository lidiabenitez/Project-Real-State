```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
target = pd.read_csv('T2018.csv')
target1 = pd.read_csv('T2019.csv')
target2 = pd.read_csv('T2020.csv')
target3 = pd.read_csv('T2021.csv')
```

```python
value1 = pd.read_csv('V2018.csv')
value2 = pd.read_csv('V2019.csv')
value3 = pd.read_csv('V2020.csv')
value4 = pd.read_csv('V2021.csv')
rent1= pd.read_csv('R2018.csv')
rent2= pd.read_csv('R2019.csv')
rent3= pd.read_csv('R2020.csv')
rent4= pd.read_csv('R2021.csv')
```

```python
merged_2 = pd.merge(left=merged_1, right=data_t, how='outer')
merged_2
```

```python
#I will reduce age categories
def cleanOperation1(y):
    if '0-4 anys' in y:
        return "0-24 years"
    elif '5-9 anys' in y:
        return "0-24 years"
    elif '10-14 anys' in y:
        return '0-24 years'
    elif '15-19 anys' in y:
        return '0-24 years'
    elif '20-24 anys' in y:
        return '0-24 years'
    elif '25-29 anys' in y:
        return '25-39 years'
    elif '30-34 anys' in y: 
        return '25-39 years'
    elif '35-39 anys' in y:
        return '25-39 years'
    elif '40-44 anys ' in y:
        return '40-64 years'
    elif '45-49 anys' in y:
        return '40-64 years'
    elif '50-54 anys' in y:  
        return '40-64 years'
    elif '55-59 anys' in y:     
        return '40-64 years'
    elif '60-64 anys' in y:     
        return '40-64 years'
    elif '65-69 anys' in y:   
        return '65-89 years'
    elif '70-74 anys' in y:     
        return '65-89 years'
    elif '75-79 anys' in y:  
        return '65-89 years'
    elif '80-84 anys' in y:
        return '65-89 years'
    elif '85-89 anys' in y:
        return '65-89 years'
    else:
        return '+95 years'
```

```python
merged_2['Edat_quinquennal'] = list(map(cleanOperation1, merged_2['Edat_quinquennal']))
```

```python
merged_2["Sexe"].value_counts()
```

```python
#same with gender
def cleansex(x):  
    if 'Home' in x:
        return 'Male'
    elif 'Dona' in x:
        return 'Female'
    if 'Homes' in x:
        return 'Male'
    elif 'Dones' in x:
        return 'Female'
    
    else:
        return x
merged_2['Sexe'] = list(map(cleansex, merged_2['Sexe']))
merged_2.head()
```

```python
#I'm going to pass it to thousands
def cleanmillers(row):
    x=row['Valor']
    y=row['Preu_mitja_habitatge']
    if 'Milers' in y:
        return round(x*1000)
    else:
        return round(x)

realstate['Valor'] = realstate.apply(cleanmillers, axis=1)
#Check
```
